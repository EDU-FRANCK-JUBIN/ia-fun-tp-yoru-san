```python
try:
    from PIL import Image
except ImportError:
        import Image
        
import pytesseract 
from pytesseract import Output
from matplotlib import pyplot as plt
import cv2
```

```python
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'

simage = r'./tesseract/image_4.png'
image = cv2.imread(simage)
print(pytesseract.image_to_string(image, lang='fra'))
plt.imshow(image, 'gray')
 
```

```python
def remove_noise(img):
    return cv2.medianBlur(img, 5)

retouche = remove_noise(image)
print(pytesseract.image_to_string(image, lang='fra'))
plt.imshow(retouche)
```

```python
def thresholding(img):
    return cv2.threshold(img, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)[1]

def grayscale(img):
    return cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

retouche2 = thresholding(grayscale(remove_noise(image)))
print(pytesseract.image_to_string(retouche2, lang='fra'))
plt.imshow(retouche2, 'gray')
```
